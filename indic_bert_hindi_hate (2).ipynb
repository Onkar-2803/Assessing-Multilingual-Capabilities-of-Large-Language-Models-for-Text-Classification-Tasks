```python
# train = train.rename(columns={"polarity": "label"}, errors="raise")
train = train.rename(columns={"label": "labels"}, errors="raise")
```

```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs

model_args = ClassificationArgs()
model_args.num_train_epochs = 1
# model_args.learning_rate = 1e-5
model_args.train_batch_size = 16
model_args.labels_list = ["Hate", "Non-Hate"]
model_args.max_seq_length = 512
model_args.overwrite_output_dir = True
# Add more configuration options as needed

model = ClassificationModel(
    "albert", "ai4bharat/indic-bert", num_labels = 2, args=model_args, use_cuda=True
)
```

```python
train.columns
```

```python
model.train_model(train)
```

```python
test = pd.read_csv('/content/test-hindi-final.csv')
```

```python
predictions, raw_outputs = model.predict(test["text"].values.tolist())
```

```python
predictions
```

```python
from sklearn.metrics import classification_report
```

```python
labels = test['label']
```
