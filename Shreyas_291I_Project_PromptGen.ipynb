```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

def calculate_metrics(actual, predicted):
    accuracy = accuracy_score(actual, predicted)
    precision = precision_score(actual, predicted, average='macro')
    recall = recall_score(actual, predicted, average='macro')
    f1_macro = f1_score(actual, predicted, average='macro')
    f1_micro = f1_score(actual, predicted, average='micro')

    print(accuracy, precision, recall, f1_macro, f1_micro)
```

```python
def encode_hate(x):
  if "non" in str(x).lower():
    return 0
  return 1

def encode_senti(x):
  if "neutral" in str(x).lower():
    return 0
  elif "positive" in str(x).lower():
    return 1
  return -1
```

```python
def do_it(path, sheet):
  df_path = pd.read_excel(path, sheet_name=sheet)
  df_path["Predicted_H"] = df_path["Predicted_H"].apply(lambda x:encode_hate(x))
  df_path["Predicted_S"] = df_path["Predicted_S"].apply(lambda x:encode_senti(x))
  df_path["Actual_S"] = df_path["Actual_S"].apply(lambda x:encode_senti(x))

  calculate_metrics(df_path["Predicted_H"].values.tolist(), df_path["Actual_H"].values.tolist())
  calculate_metrics(df_path["Predicted_S"].values.tolist(), df_path["Actual_S"].values.tolist())
```
