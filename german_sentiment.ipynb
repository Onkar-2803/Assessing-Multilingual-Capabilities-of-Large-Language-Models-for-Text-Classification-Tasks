```python
prompts = []

for idx in range(0, 100, 5):
    prompts.append(f"""You are a text classification model. You have to classify data for a multiclass hate classification task for German. 
The sentence can be classified into 2 different categories namely: Positive, Negative. 

I'll give 2 examples of each of the category, and then pass 5 sentences. You'll need to classify the sentences into one of the above mentioned labels.

Negative: {trainSet['text'][0]}
Negative: {trainSet['text'][1]}
Positive: {trainSet['text'][2]}
Positive: {trainSet['text'][3]}

Classify following sentences.

{idx}. {test_final['text'][idx]}
{idx+1}. {test_final['text'][idx+1]}
{idx+2}. {test_final['text'][idx+2]}
{idx+3}. {test_final['text'][idx+3]}
{idx+4}. {test_final['text'][idx+4]}
""")
```

```python
promptsDF = pd.DataFrame()
promptsDF['prompt'] = prompts

promptsDF.to_csv('prompts-german-sentiment.csv', index=False)
```
